```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# prompt: read housing file , the seperate here is not csv its spaces
data = pd.read_csv('housing.csv', delim_whitespace=True)
data
```

```python
data.isnull().sum()
```

```python
data.describe
```

```python
X = data.drop('24.00', axis = 1)
y = data['24.00']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test
```

```python
# prompt: linear regression and name it model 1 with r2 score for train and test

model1 = LinearRegression()
model1.fit(X_train, y_train)

y_train_pred = model1.predict(X_train)
y_test_pred = model1.predict(X_test)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f"R2 score on training data: {r2_train:.4f}")
print(f"R2 score on testing data: {r2_test:.4f}")
```

```python
# prompt: polynomial featrue

poly_features = PolynomialFeatures(degree=2)

X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

model2 = LinearRegression()
model2.fit(X_train_poly, y_train)

y_train_pred_poly = model2.predict(X_train_poly)
y_test_pred_poly = model2.predict(X_test_poly)

r2_train_poly = r2_score(y_train, y_train_pred_poly)
r2_test_poly = r2_score(y_test, y_test_pred_poly)

print(f"R2 score with Polynomial Features (degree 2) on training data: {r2_train_poly:.4f}")
print(f"R2 score with Polynomial Features (degree 2) on testing data: {r2_test_poly:.4f}")
```

```python
# prompt: lasso regression on polynomial model

from sklearn.linear_model import Lasso

# Lasso regression with polynomial features
lasso_model = Lasso(alpha=0.1) # You can adjust the alpha parameter

lasso_model.fit(X_train_poly, y_train)

y_train_pred_lasso = lasso_model.predict(X_train_poly)
y_test_pred_lasso = lasso_model.predict(X_test_poly)

r2_train_lasso = r2_score(y_train, y_train_pred_lasso)
r2_test_lasso = r2_score(y_test, y_test_pred_lasso)

print(f"\nR2 score with Lasso Regression (Polynomial Features) on training data: {r2_train_lasso:.4f}")
print(f"R2 score with Lasso Regression (Polynomial Features) on testing data: {r2_test_lasso:.4f}")
```

```python
# prompt: ridge regression

from sklearn.linear_model import Ridge

# Ridge regression with polynomial features
ridge_model = Ridge(alpha=0.1) # You can adjust the alpha parameter

ridge_model.fit(X_train_poly, y_train)

y_train_pred_ridge = ridge_model.predict(X_train_poly)
y_test_pred_ridge = ridge_model.predict(X_test_poly)

r2_train_ridge = r2_score(y_train, y_train_pred_ridge)
r2_test_ridge = r2_score(y_test, y_test_pred_ridge)

print(f"\nR2 score with Ridge Regression (Polynomial Features) on training data: {r2_train_ridge:.4f}")
print(f"R2 score with Ridge Regression (Polynomial Features) on testing data: {r2_test_ridge:.4f}")
```
