```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# prompt: load diabest regression data set

diabetes = load_diabetes()
diabetes
```

```python
x = diabetes.data
y = diabetes.target
print(x.shape, y.shape)
```

```python
x = pd.DataFrame(x)
y = pd.DataFrame(y)
x
```

```python
x.isnull().sum()
```

```python
# prompt: split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
x_test
```

```python
model_1 = linear_model.LinearRegression()
model_1.fit(x_train, y_train)
```

```python
y_train_pred = model_1.predict(x_train) #check for under fitting
# print the mean square error for the train data
print("Mean squared error for train data: ", mean_squared_error(y_train, y_train_pred))
```

```python
# print the r2 score for the train data

print("R2 score for train data: ", r2_score(y_train, y_train_pred))
```

```python
y_test_pred = model_1.predict(x_test) #check for over fitting
# print the mean square error for the test data

print("Mean squared error for test data: ", mean_squared_error(y_test, y_test_pred))
```

```python
# print the r2 score for the test data

print("R2 score for test data: ", r2_score(y_test, y_test_pred))
```

```python
# prompt: polynomial featrue


poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

model_2 = linear_model.LinearRegression()
model_2.fit(x_train_poly, y_train)

y_train_pred_poly = model_2.predict(x_train_poly)
print("Mean squared error for train data with polynomial features: ", mean_squared_error(y_train, y_train_pred_poly))
print("R2 score for train data with polynomial features: ", r2_score(y_train, y_train_pred_poly))

y_test_pred_poly = model_2.predict(x_test_poly)
print("Mean squared error for test data with polynomial features: ", mean_squared_error(y_test, y_test_pred_poly))
print("R2 score for test data with polynomial features: ", r2_score(y_test, y_test_pred_poly))
```

```python
# prompt: lasso regression on polynomial model

# Implement Lasso regression on the polynomial features
model_lasso = linear_model.Lasso(alpha=0.1) # You can tune the alpha parameter

# Fit the Lasso model to the polynomial features
model_lasso.fit(x_train_poly, y_train)

# Make predictions on the training and testing sets
y_train_pred_lasso = model_lasso.predict(x_train_poly)
y_test_pred_lasso = model_lasso.predict(x_test_poly)

# Evaluate the Lasso model
print("\nLasso Regression with Polynomial Features:")
print("Mean squared error for train data: ", mean_squared_error(y_train, y_train_pred_lasso))
print("R2 score for train data: ", r2_score(y_train, y_train_pred_lasso))
print("Mean squared error for test data: ", mean_squared_error(y_test, y_test_pred_lasso))
print("R2 score for test data: ", r2_score(y_test, y_test_pred_lasso))
```

```python
# prompt: ridge regression

# Implement Ridge regression on the polynomial features
model_ridge = linear_model.Ridge(alpha=0.1) # You can tune the alpha parameter

# Fit the Ridge model to the polynomial features
model_ridge.fit(x_train_poly, y_train)

# Make predictions on the training and testing sets
y_train_pred_ridge = model_ridge.predict(x_train_poly)
y_test_pred_ridge = model_ridge.predict(x_test_poly)

# Evaluate the Ridge model
print("\nRidge Regression with Polynomial Features:")
print("Mean squared error for train data: ", mean_squared_error(y_train, y_train_pred_ridge))
print("R2 score for train data: ", r2_score(y_train, y_train_pred_ridge))
print("Mean squared error for test data: ", mean_squared_error(y_test, y_test_pred_ridge))
print("R2 score for test data: ", r2_score(y_test, y_test_pred_ridge))
```
